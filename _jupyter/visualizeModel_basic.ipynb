```r
# import libraries
library(dplyr)
library(ggplot2)
library(tidyr)

# import data
df <- read.table('../data_examples/MAX101_Main_block_MR_I.x1D')
```

```r
# subset dataframe to only include model regressors & label

df.long <- df %>% gather()
timepoints <- length(df.long$key)/length(unique(df.long$key))

df.long <- df.long %>%
           mutate(TR = rep(1:timepoints, 57)) %>%
           mutate(index = rep(1:57, each = timepoints))

df.model <- df.long %>%
            filter(index %in% (25:45)) %>% mutate(key = replace(key, key=="V45", "falsePosThreat-seed")) %>%
             mutate(key = replace(key, key=="V44", "falsePosSafe-seed")) %>%
             mutate(key = replace(key, key=="V43", "falseNeuThreat-seed")) %>%
             mutate(key = replace(key, key=="V42", "falseNeuSafe-seed")) %>%
             mutate(key = replace(key, key=="V41", "seed")) %>%
             mutate(key = replace(key, key=="V40", "respond-realPosThreat")) %>%
             mutate(key = replace(key, key=="V39", "respond-realPosSafe")) %>%
             mutate(key = replace(key, key=="V38", "respond-realNeuThreat")) %>%
             mutate(key = replace(key, key=="V37", "respond-realNeuSafe")) %>%
             mutate(key = replace(key, key=="V36", "respond-falsePosThreat")) %>%
             mutate(key = replace(key, key=="V35", "respond-falsePosSafe")) %>%
             mutate(key = replace(key, key=="V34", "respond-falseNeuThreat")) %>%
             mutate(key = replace(key, key=="V33", "respond-falseNeuSafe")) %>%
             mutate(key = replace(key, key=="V32", "realPosThreat")) %>%
             mutate(key = replace(key, key=="V31", "realPosSafe")) %>%
             mutate(key = replace(key, key=="V30", "realNeuThreat")) %>%
             mutate(key = replace(key, key=="V29", "realNeuSafe")) %>%
             mutate(key = replace(key, key=="V28", "falsePosThreat")) %>%
             mutate(key = replace(key, key=="V27", "falsePosSafe")) %>%
             mutate(key = replace(key, key=="V26", "falseNeuThreat")) %>%
             mutate(key = replace(key, key=="V25", "falseNeuSafe")) 
```

```r
# Re-order label values with regressor names
df.model$key <- factor(df.model$key, levels = c("seed",
                                                "falsePosThreat-seed",
                                                "falsePosThreat",
                                                "falsePosSafe-seed",
                                                "falsePosSafe",
                                                "falseNeuThreat-seed",
                                                "falseNeuThreat",
                                                "falseNeuSafe-seed",
                                                "falseNeuSafe",
                                                "realPosThreat",
                                                "realPosSafe",
                                                "realNeuThreat",
                                                "realNeuSafe",
                                                "respond-realPosThreat",
                                                "respond-realPosSafe",
                                               "respond-realNeuThreat",
                                               "respond-realNeuSafe",
                                               "respond-falsePosThreat",
                                               "respond-falsePosSafe",
                                               "respond-falseNeuThreat",
                                               "respond-falseNeuSafe"))
```

## Model regressors

```r
ggplot(df.model, aes(x = TR, y = value)) + 
  geom_line(aes(color = index), size = 1) +
  facet_grid(rows = vars(key), scales = "free") +
  scale_x_continuous(breaks = c(0, 340, 680, 1020, 1360, 1700, 2040)) +
  scale_y_continuous() +
  theme_minimal() +
  labs(y = NULL, title = "Regressors") +
  theme(
        legend.position = 'none',
        strip.text.y = element_text(angle = 0, hjust = 0),
        axis.text.y = element_text(size = 5))

ggsave('../assets/images/MAX_basicModel_gPPI.png')
```
