## Look at time series 252 TRs at a time
### Compare cerebellum region of interest with anatomical V1 timeseries

One participant's extracted timeseries

```r
library(tidyverse)

#import cerebellum timeseries (l-Crus II) & V1
df <- read.table('/Users/kellymorrow/gPPI/output/MAX101_l-CrusII_seed_NoBaseline_avg.1D', col.names = c("value"))
v1 <- read.table('/Users/kellymorrow/gPPI/output/MAX101_l-BA17_seed_NoBaseline_avg.1D', col.names = c("value"))

nrow <- length(df$value)
rep <- nrow/8
labs <- factor(c("TR 0-252","TR 253-505","TR 506-758","TR 759-1011","TR 1012-1264","TR 1265-1517","TR 1518-1770","TR 1771-2016"),
              levels = c("TR 0-252","TR 253-505","TR 506-758","TR 759-1011","TR 1012-1264","TR 1265-1517","TR 1518-1770","TR 1771-2016"))
```

```r
func <- function(df){
    df %>% mutate(timepoint = rep(1:rep, 8)) %>%
    mutate(group = rep(labs, each = (nrow/8))) %>%
    ggplot(aes(x = timepoint, y = value)) + 
    facet_wrap(~group, scales = "free") +
    geom_line() +
    theme(
    aspect.ratio = .5)
}    
```

### Cerebellum seed timeseries with baseline removed
Baseline is subtracted from functional data. Each plot is 252 timepoints to get a better look at the raw data. 

```r
func(df)
```

### V1 time series with baseline removed
This is for comparison to cerebellum timeseries

```r
func(v1)
```
